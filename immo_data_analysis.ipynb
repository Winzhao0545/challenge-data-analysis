```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file=r'C:\Users\frede_0021xgx\OneDrive\Documents\GitHub\ImmoEliza\data_output\dataframe.csv'
immo_info=pd.read_csv(file)
del immo_info['Unnamed: 0']
immo_info
```

```python
#remove repeated rows
immo_info_drop_dup=immo_info.drop_duplicates()
#immo_info_drop_dup
```

```python
#replace negative number with zero,df[df < 0] = 0
num=immo_info_drop_dup._get_numeric_data()
num[num<0]=0

immo_info_drop_dup
```

```python
#replace 'UNKNOWN' value with zero
immo_info_drop_dup=immo_info_drop_dup.replace('UNKNOWN',0)
immo_info_drop_dup
```

```python
#percentage of missing value
percent_m_v=immo_info_drop_dup.isna().sum()/17047

#Plot of missing value
percent_m_v.plot(kind='bar')
plt.show()
```

```python
#delete rows with mising price 
immo_all_price=immo_info_drop_dup.dropna(subset='price')
#immo_all_price.head()
```

```python
immo_all_price['locality']=immo_all_price['locality'].str.capitalize()
immo_all_price
```

```python
#show percentage of proporty number in different province
props_region=immo_all_price['region'].value_counts().rename_axis('reg_count').reset_index(name='counts')
props_region
labels=props_region.reg_count
value=props_region.counts

plt.pie(value, labels=labels, autopct='%1.0f%%')

plt.axis('equal')
plt.title('Propety proportion of different region')

plt.show()
```

```python
immo_reg_pro_avg=immo_all_price.groupby(['region','province']).agg('price').mean().sort_values(ascending=False)
immo_reg_pro_avg=immo_reg_pro_avg.reset_index()
sns.barplot(data=immo_reg_pro_avg, x='price', y="province",hue='region').set(title='Average property price by region and province')
```

```python
# subset for Region, province,locality and price to keep
immo_im_info = immo_all_price[['region','province','locality','property_type','price']]
immo_im_info
```

```python
# subset the df for one region
immo_reg=immo_im_info.set_index(['region'])
keep_bru=['Brussels']
immo_reg_bru=immo_reg.loc[keep_bru]
immo_reg_bru
```

```python
# reset top_bru
immo_reg_bru=immo_reg_bru.reset_index()
immo_reg_bru
```

```python
#Top 5 expenive municipalities in Brussels
immo_reg_bru_avg_most = immo_reg_bru.groupby(['region','province','locality','property_type']).agg('price').mean().sort_values(ascending=False)
immo_reg_bru_avg_most= immo_reg_bru_avg_most.reset_index().head(5)
display(immo_reg_bru_avg_most)
```

```python
#set color
palette ={"APARTMENT": "C0", "HOUSE": "C1"}
# Plot of relation between Region and Property_Type
sns.barplot(data=immo_reg_bru_avg_most, x='price', y="locality",hue='property_type',palette=palette).set(title='Most expensive municipalities in Brussels')
```

```python
# less expensive municipalities in Brussels
immo_reg_bru_avg_less = immo_reg_bru.groupby(['region','province','locality','property_type']).agg('price').mean().sort_values()
immo_reg_bru_avg_less= immo_reg_bru_avg_less.reset_index().head(5)
immo_reg_bru_avg_less

#set color for property type
palette ={"APARTMENT": "C0", "HOUSE": "C1"}
sns.barplot(data=immo_reg_bru_avg_less, x='price', y="locality",hue='property_type',palette=palette).set(title='Less expensive Municipalities in Brussels')
```

```python
#correlation between price and living area
 #keep living area that less than 1000
immo_liv=immo_all_price[immo_all_price['living_area']<600.00]
immo_liv=immo_liv.groupby('living_area').agg('price').mean().reset_index()
#regression line
sns.regplot(x = "living_area", y = "price",data=immo_liv,
scatter_kws = {"color": "black", "alpha": 0.5},line_kws = {"color": "red"}).set(title='Correlation between price and living area')
```

```python
#correlation between differernt variables
immo_liv['price'].corr(immo_liv['living_area'])

```
