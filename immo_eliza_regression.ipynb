```python
'''Data cleaning'''
#Import neccessary liberies 
import pandas as pd
```

```python
# read data
file=r'C:\Users\frede_0021xgx\OneDrive\Documents\GitHub\challenge-data-analysis\data\dataframe_cleaned.csv'
immo_info=pd.read_csv(file)
```

```python
#data cleaning
#1.drop duplicates
immo_info=immo_info.drop_duplicates()
#2.No NaNs(replace NaNs with 0)
immo_info=immo_info.fillna(0)
#3.No text data(replace 'UNKNOWN' value with zero)
immo_info=immo_info.replace('UNKNOWN',0)
#4.subset the data positive value  
    #price  
immo_info=immo_info[immo_info['price']>0.0]
    #number_rooms 
immo_info=immo_info[immo_info['number_rooms']>0.0]
    #terrace_area 
immo_info=immo_info[immo_info['terrace_area']>0.0]
    #living_area
immo_info=immo_info[immo_info['living_area']>0.0]
    #garden_area
immo_info=immo_info[immo_info['garden_area']>0.0]
    #surface_land  
immo_info['surface_land']=immo_info['surface_land'].astype(float)
immo_info=immo_info[immo_info['surface_land']>0.0]
immo_info
```

```python
''' Data formating'''
#Import neccessary liberies 
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from matplotlib import pyplot as plt
```

```python
#Convert dataframe to numpy array
X=immo_info[['number_rooms','terrace_area','living_area','garden_area','surface_land']].to_numpy()
y=immo_info['price'].to_numpy()
X.shape,y.shape
```

```python
# Set the shape of array to fit into the model
#X=X.reshape(-1,1)
#y=y.reshape(-1,1)
#Split dataset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#Fit the trained data into a model(linear regression)
regressor=LinearRegression()
#train model with data
regressor.fit(X_train, y_train)
#display score of model
print(regressor.score(X_train, y_train))
```
