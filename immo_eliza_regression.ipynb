```python
'''Data cleaning'''
#Import neccessary liberies 
import pandas as pd
```

```python
# read data
file=r'C:\Users\frede_0021xgx\OneDrive\Documents\GitHub\challenge-data-analysis\data\dataframe_cleaned.csv'
immo_info=pd.read_csv(file)
```

```python
#data cleaning
#1.drop duplicates
immo_info=immo_info.drop_duplicates()
#2.No NaNs(replace NaNs with 0)
immo_info=immo_info.fillna(0)
#3.No text data(replace 'UNKNOWN' value with zero)
immo_info=immo_info.replace('UNKNOWN',0)
immo_info
```

```python
''' Data formating'''
#Import neccessary liberies 
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from matplotlib import pyplot as plt
```

```python
#Convert dataframe to numpy array
X=immo_info[['number_rooms','garden_area','terrace_area','surface_land','swimming_pool','swimming_pool' ]].to_numpy()
y=immo_info['price'].to_numpy()
X.shape,y.shape
```

```python
# Set the shape of array to fit into the model
#X=X.reshape(-1,1)
#y=y.reshape(-1,1)
#Split dataset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#Fit the trained data into a model(linear regression)
regressor=LinearRegression()
#train model with data
regressor.fit(X_train, y_train)
#display score of model
print(regressor.score(X_train, y_train))
```
